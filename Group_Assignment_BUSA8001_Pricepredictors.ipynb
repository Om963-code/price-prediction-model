```python
# Extracting numerical values from 'back_legroom'
combined_df['back_legroom'] = combined_df['back_legroom'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'front_legroom'
combined_df['front_legroom'] = combined_df['front_legroom'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'fuel_tank_volume'
combined_df['fuel_tank_volume'] = combined_df['fuel_tank_volume'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'height'
combined_df['height'] = combined_df['height'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'length'
combined_df['length'] = combined_df['length'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'maximum_seating'
combined_df['maximum_seating'] = combined_df['maximum_seating'].str.extract(r'(\d+)').astype(float)

# Extracting numerical values from 'power'
combined_df['power'] = combined_df['power'].str.extract(r'(\d+)').astype(float)

# Extracting numerical values from 'torque'
combined_df['torque'] = combined_df['torque'].str.extract(r'(\d+)').astype(float)

# Extracting numerical values from 'wheelbase'
combined_df['wheelbase'] = combined_df['wheelbase'].str.extract(r'(\d+\.*\d*)').astype(float)

# Extracting numerical values from 'width'
combined_df['width'] = combined_df['width'].str.extract(r'(\d+)').astype(float)
```

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, map colours into 5 basic colours + 'other': [red, yellow, green, blue, purple, other] and then encode.  
(2 marks)

```python
## Task 2, Question 4 Code Here
from sklearn.preprocessing import OneHotEncoder

# Updated categorical columns in the combined_df
categorical_columns_combined = ['body_type', 'city', 'engine_type', 'exterior_color', 'fuel_type', 'interior_color', 'make_name', 'model_name', 'transmission', 'transmission_display', 'wheel_system', 'listing_color']

for col in categorical_columns_combined:
    # Encode 'other' for observations with multiple values
    combined_df[col] = combined_df[col].apply(lambda x: x if len(str(x).split()) == 1 else 'other')
    # Encode the top 5 values and 'other' for columns with more than 5 unique values
    if len(combined_df[col].unique()) > 5:
        top_5 = combined_df[col].value_counts().index[:5]
        combined_df[col] = combined_df[col].apply(lambda x: x if x in top_5 else 'other')
    
    # Fit and transform the OneHotEncoder
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    encoded_cols = pd.DataFrame(encoder.fit_transform(combined_df[[col]]))
    
    # Rename columns based on categories
    encoded_cols.columns = [col + '_' + str(i) for i in range(encoded_cols.shape[1])]
    
    # Concatenate with the original dataframe
    combined_df = pd.concat([combined_df, encoded_cols], axis=1)

    # Drop the original categorical columns
    combined_df.drop(col, axis=1, inplace=True)

combined_df.info()
```

**Explanation:**

**Categorical Columns Selection:** The code first defines a list categorical_columns_combined containing the names of categorical columns in the combined_df DataFrame.

**Handling Multiple Values:** The code checks if a categorical variable contains multiple values for an observation. If it does, it is encoded as 'other'.

**Top 5 Values Encoding:** For categorical features with more than 5 unique values, the code selects the top 5 most frequent values and encodes them, along with 'other'.

**One-Hot Encoding:** The OneHotEncoder from scikit-learn is used to transform the categorical variables into binary vectors (one-hot encoding). The handle_unknown='ignore' parameter ensures that any previously unseen categories during transformation are ignored.

**Column Renaming and Concatenation:** The resulting encoded columns are given appropriate names based on the original categorical variable names and index values. They are then concatenated with the original DataFrame.

**Drop Original Categorical Columns:** The original categorical columns are dropped from the DataFrame after encoding to avoid redundancy.
